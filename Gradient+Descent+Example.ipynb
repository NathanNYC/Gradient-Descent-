```python
next_x = 6 # start search @ 6
gamma = 0.01 # step size multiplier
precision = 0.00001 # result precision
max_iters = 100 # number of iterations
iters=0

def df(x): # derivative function -> whatever is returned  
    #return 4 * x ** 3 - 9 * x ** 2
    return 2*(x+5)

for _ in range (max_iters):
    current_x = next_x
    next_x = current_x - gamma * df(current_x)

    
    step = next_x - current_x
    iters+=1
    print("Iteration:",iters,"\n X value is",round(current_x,3))
    if abs(step) <= precision:
        break
        
print("Minimum at ", next_x)
```
